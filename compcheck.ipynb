```python
from csv import writer
from csv import reader
default_text1 = 'phcompliant'
default_text2 = 'pahcompliant'
default_text3 = 'turbcompliant'
default_text4 = 'gascompliant'

with open('/Users/91893/Downloads/output_2.csv', 'r') as read_obj, \
        open('/Users/91893/Documents/synergy/output_3.csv', 'w', newline='') as write_obj:
    csv_reader = reader(read_obj)
    csv_writer = writer(write_obj)

    for row in csv_reader:
        row.append(default_text1)
        row.append(default_text2)
        row.append(default_text3)
        row.append(default_text4)
        csv_writer.writerow(row)
```

```python
import pandas as pd
from pandas import *
import math
cs = read_csv('/Users/91893/Documents/synergy/output_3.csv')
ph = cs['dischargePh'].tolist()
pah = cs['dischargePah'].tolist()
tur = cs['turbidityDifference'].tolist()
gratio = cs['gasRatio'].tolist()
eca = cs['0'].tolist()

csf = pd.read_csv('/Users/91893/Documents/synergy/output_3.csv')
#ph
for i in range (len(ph)):
    print(i)
    if math.isnan(ph[i]):
        csf.loc[i,'phcompliant']=''
        csf.to_csv('/Users/91893/Documents/synergy/output_3.csv',index=False)
    elif(ph[i]<3.03):
        csf.loc[i,'phcompliant']='0'
        csf.to_csv('/Users/91893/Documents/synergy/output_3.csv',index=False)
    else:
        csf.loc[i,'phcompliant']='1' 
        csf.to_csv('/Users/91893/Documents/synergy/output_3.csv',index=False)
#gasratio
    if math.isnan(gratio[i]):
        csf.loc[i,'gascompliant']=''
        
    else:
        if eca[i]==0:
            if gratio[i]<21.7:
                csf.loc[i,'gcompliant']='1'
                csf.to_csv('/Users/91893/Documents/synergy/output_3.csv',index=False)
            else:
                csf.loc[i,'gcompliant']='0'
                csf.to_csv('/Users/91893/Documents/synergy/output_3.csv',index=False)
        else:
            if gratio[i]<4.3:
                csf.loc[i,'gcompliant']='1'
                csf.to_csv('/Users/91893/Documents/synergy/output_3.csv',index=False)
            else:
                csf.loc[i,'gcompliant']='0'
                csf.to_csv('/Users/91893/Documents/synergy/output_3.csv',index=False)
            

#pah
for i in range(len(pah)-480):
    c=0
    ct=0
    print (i,'*')
    for j in range(480):
        if math.isnan(pah[i+j]):
            continue
        elif pah[i+j]>(51.9*2):
            c=c+1
    if math.isnan(pah[i]):
        csf.loc[i,'pahcompliant']=''
        csf.to_csv('/Users/91893/Documents/synergy/output_3.csv',index=False)
    elif c>(15*60/90):
        csf.loc[i,'pahcompliant']='0'
        csf.to_csv('/Users/91893/Documents/synergy/output_3.csv',index=False)
    else:
        csf.loc[i,'pahcompliant']='1'   
        csf.to_csv('/Users/91893/Documents/synergy/output_3.csv',index=False)

#turbidity

    for j in range(480):
        if math.isnan(tur[i+j]):
            continue
        elif tur[i+j]>1.5*25:
            ct=ct+1
    if math.isnan(tur[i]):
            csf.loc[i,'turbcompliant']=''
            csf.to_csv('/Users/91893/Documents/synergy/output_3.csv',index=False)
    elif (ct*90)>15*60:  #non compliant
        csf.loc[i,'turbcompliant']='0'
        csf.to_csv('/Users/91893/Documents/synergy/output_3.csv',index=False)
    else:
        csf.loc[i,'turbcompliant']='1'
        csf.to_csv('/Users/91893/Documents/synergy/output_3.csv',index=False)
```
